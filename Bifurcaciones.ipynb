# **Ejercicios Bifurcaciones**

Módulos de librerias

```
import matplotlib.style
matplotlib.style.use('classic')
import numpy as np
import time
from matplotlib import pyplot as plt
from scipy.integrate import solve_ivp
```

## Ejercicio 3.4.5

\begin{equation}
 \frac{dx}{dt}=-x+rtanh(x)
 \end{equation} 

```
x=np.arange(-10, 10, 0.01)

r=2
dxdt= -x+r*np.tanh(x)
fig = plt.figure()
plt.plot( x, dxdt, linewidth = 2)
plt.title(r'Gráfico $\frac{dx}{dt}$ vs $x$', fontsize=20)
plt.xlim([-11, 11])
#plt.ylim([2.5, 5.5])
plt.xlabel(r'$x$', fontsize=20)
plt.ylabel(r'$\frac{dx}{dt}$', fontsize=20)
plt.grid()
plt.show()
```

```
def sistema_2(t, x,a,b):
    x_dot = -x+a*np.tanh(x)
    return x_dot
```

```
b = 4,3
x0 = np.linspace(-6,6,5)
tspan = [0, 100]
r = np.linspace(0,4,50)
tic = time.time()
for j in range(len(r)):
    sol = solve_ivp(sistema_2, tspan, x0,args=(r[j],b))
    t = sol.t
    x = sol.y
    for i in range(x0.size):
        plt.scatter(r[j],x[i,-1] , color='green', marker='.')

x0 = np.linspace(-1,0,5)
tspan = [0, -100]

r = np.linspace(1.5,4,100)
tic = time.time()
for j in range(r.size):
    sol = solve_ivp(sistema_2, tspan, x0,args=(r[j],b))
    t = sol.t
    x = sol.y
    for i in range(x0.size):
        plt.scatter(r[j],x[i,-1] , color='red', marker='.')
        plt.title('Diagrama de bifurcación')
        plt.xlabel('r')
        plt.ylabel('x')

toc = time.time()
plt.show()
```

## Ejercicio 3.4.4
\begin{equation}
 \frac{dx}{dt}=rx-sinh(x)
 \end{equation} 

```
x=np.arange(-10, 10, 0.01)

dxdt= 2*x-np.sinh(x)
fig = plt.figure()
plt.plot( x, dxdt, linewidth = 2)
plt.title(r'Gráfico $\frac{dx}{dt}$ vs $x$', fontsize=20)
plt.xlim([-11, 11])
#plt.ylim([2.5, 5.5])
plt.xlabel(r'$x$', fontsize=20)
plt.ylabel(r'$\frac{dx}{dt}$', fontsize=20)
plt.grid()
plt.show()
```

```
def sistema_2(t, x,a,b):
    x_dot = a*x-np.sinh(x)
    return x_dot
```

```
b = 4,3
x0 = np.linspace(-10,10,10)
tspan = [0, 100]
r = np.linspace(-5,5,100)
tic = time.time()
for j in range(len(r)):
    sol = solve_ivp(sistema_2, tspan, x0,args=(r[j],b))
    t = sol.t
    x = sol.y
    for i in range(x0.size):
        plt.scatter(r[j],x[i,-1] , color='green', marker='.')

x0 = np.linspace(-.1,.1,5)
tspan = [0, -100]

r = np.linspace(1.1,5,50)
tic = time.time()
for j in range(r.size):
    sol = solve_ivp(sistema_2, tspan, x0,args=(r[j],b))
    t = sol.t
    x = sol.y
    for i in range(x0.size):
        plt.scatter(r[j],x[i,-1] , color='red', marker='.')
        plt.title('Diagrama de bifurcación')
        plt.xlabel('r')
        plt.ylabel('x')

toc = time.time()
plt.show()
```
