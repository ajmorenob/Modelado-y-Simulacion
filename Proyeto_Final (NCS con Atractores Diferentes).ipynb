##Un Sistema Caótico Simple con Atractores Topológicamente Diferentes

Después del descubrimiento del conocido sistema de Lorenz, Las últimas décadas han visto muchos documentos discutiendo modelos caóticos deterministas . Atractores caóticos son
caracterizado por características como una fuerte dependencia de las condiciones iniciales, sensibilidad a los cambios en el sistema, parámetros, la presencia de armónicos fuertes en las señales, la dimensión fraccional de las trayectorias del estado espacial, y la presencia de direcciones de estiramiento representadas por un positivo Exponente de Lyapunov.

NUEVO SISTEMA CAÓTICO (NCS)

La matemática modelo toma la forma:
sistema (1)

\begin{equation}
 \dot{x}=ax-yz,
\end{equation} 
\begin{equation}
 \dot{y}=-by+xz,
\end{equation} 
\begin{equation}
 \dot{z}=c-z+xy
\end{equation} 
 

 El sistema (1) proporciona un atractor autoexcitado con cinco puntos de equilibrio, que es topológicamente diferente a otros
sistemas bien conocidos como los sistemas de Lorenz y Rossler,
que tienen tres y dos equilibrios respectivamente. en este papel
el sistema (1) se modifica mediante la inclusión de un sistema no lineal adicional
término en la ecuación del tercer estado para z, generando así seis
diferentes atractores caóticos autoexcitados, como se muestra en la Tabla 1(PRESENTADA EN EL ARTICULO ORIGINAL).
El modelo matemático del nuevo sistema propuesto es
\begin{equation}
 \dot{x}=ax-yz,
\end{equation} 
\begin{equation}
 \dot{y}=-by+xz,
\end{equation} 
\begin{equation}
 \dot{z}=c-z+xy+dzy
\end{equation} 
 donde a, b, c, d son los parámetros del sistema. Llamamos (2)
el "NCS".
Los casos que se muestran en la Tabla 1 se derivan del sistema (2) para
diferentes valores de parámetros,
Para las condiciones iniciales [0.1,0.1,0] y los valores de los parámetros como se muestra en la Tabla. 1, se muestran los retratos de fase en la tabla 1.
en las imagenes (a-f). Se puede ver en las cifras que la propuesta
NCS tiene atractores topológicamente diferentes para los seis casos.

Módulos de librerias

```
import matplotlib.style
#matplotlib.style.use('classic')
import numpy as np
import time
from matplotlib import pyplot as plt
from scipy.integrate import solve_ivp
from mpl_toolkits.mplot3d import axes3d
```

```
def NCS(t,xyz,a,b,c,d):
    x,y,z = xyz
    return [a*x-y*z,-b*y+x*z,c-z+x*y+d*z*y]
```

```
time = [0,3000]
t = np.arange(300,3000,0.001)
init_condi = [0.3,-0.4,-0.1]
solucion = solve_ivp(NCS,time,init_condi,args=[0.2,0.1,0,0],t_eval=t)
t = solucion.t
x = solucion.y[0,:]
y = solucion.y[1,:]
z = solucion.y[2,:]

time2 = [0,6000]
t2 = np.arange(4000,6000,0.001)
init_condi2 = [0.1,0.1,0]
solucion2 = solve_ivp(NCS,time2,init_condi2,args=[0.1,0.1,0.001,0.1],t_eval=t2)
t2 = solucion2.t
x2 = solucion2.y[0,:]
y2 = solucion2.y[1,:]
z2 = solucion2.y[2,:]

time3 = [0,690]
t3 = np.arange(200,628,0.001)
init_condi3 = [0.1,0.1,0]
solucion3 = solve_ivp(NCS,time3,init_condi3,args=[0.12,0.1,1,0],t_eval=t3)
t3 = solucion3.t
x3 = solucion3.y[0,:]
y3 = solucion3.y[1,:]
z3 = solucion3.y[2,:]

time4 = [0,3000]
t4 = np.arange(1000,3000,0.001)
init_condi4 = [0.1,0.1,0]
solucion4 = solve_ivp(NCS,time4,init_condi4,args=[0.12,0.115,0.5,0],t_eval=t4)
t4 = solucion4.t
x4 = solucion4.y[0,:]
y4 = solucion4.y[1,:]
z4 = solucion4.y[2,:]

time5 = [0,3000]
t5 = np.arange(300,3000,0.001)
init_condi5 = [0.1,0.1,0]
solucion5 = solve_ivp(NCS,time5,init_condi5,args=[0.12,0.1,0,0.1],t_eval=t5)
t5 = solucion5.t
x5 = solucion5.y[0,:]
y5 = solucion5.y[1,:]
z5 = solucion5.y[2,:]

time6 = [0,1500]
t6 = np.arange(100,1500,0.001)
init_condi6 = [0.1,0.1,0]
solucion6 = solve_ivp(NCS,time6,init_condi6,args=[0.7,0.1,0.001,0.1],t_eval=t6)
t6 = solucion6.t
x6 = solucion6.y[0,:]
y6 = solucion6.y[1,:]
z6 = solucion6.y[2,:]
```

```
fig = plt.figure(figsize=(10,35))
#fig.suptitle('a',fontsize=20)


ax = fig.add_subplot(6, 2, 1)
ax.plot(-x,-y)
ax.grid(True)
ax.set_title('a',fontsize=20)
ax.set_ylabel('y')
ax.set_xlabel('x')

ax = fig.add_subplot(6, 2, 2)
ax.plot(z,-x)
ax.grid(True)
ax.set_ylabel('x')
ax.set_xlabel('z')

ax = fig.add_subplot(6, 2, 7)
ax.plot(x2,y2)
ax.grid(True)
ax.set_title('d',fontsize=20)
ax.set_ylabel('y')
ax.set_xlabel('x')

ax = fig.add_subplot(6, 2, 8)
ax.plot(z2,x2)
ax.grid(True)
ax.set_ylabel('x')
ax.set_xlabel('z')

ax = fig.add_subplot(6, 2, 3)
ax.plot(-x3,-y3)
ax.grid(True)
ax.set_title('b',fontsize=20)
ax.set_ylabel('y')
ax.set_xlabel('x')

ax = fig.add_subplot(6, 2, 4)
ax.plot(-x3,z3)
ax.grid(True)
ax.set_ylabel('x')
ax.set_xlabel('z')

ax = fig.add_subplot(6, 2, 9)
ax.plot(x4,y4)
ax.grid(True)
ax.set_title('e',fontsize=20)
ax.set_ylabel('y')
ax.set_xlabel('x')

ax = fig.add_subplot(6, 2, 10)
ax.plot(z4,x4)
ax.grid(True)
ax.set_ylabel('x')
ax.set_xlabel('z')

ax = fig.add_subplot(6, 2, 5)
ax.plot(-x5,y5)
ax.grid(True)
ax.set_title('c',fontsize=20)
ax.set_ylabel('y')
ax.set_xlabel('x')

ax = fig.add_subplot(6, 2, 6)
ax.plot(-z5,-x5)
ax.grid(True)
ax.set_ylabel('x')
ax.set_xlabel('z')

ax = fig.add_subplot(6, 2, 11)
ax.plot(x6,y6)
ax.grid(True)
ax.set_title('f',fontsize=20)
ax.set_ylabel('y')
ax.set_xlabel('x')

ax = fig.add_subplot(6, 2, 12)
ax.plot(z6,x6)
ax.grid(True)
ax.set_ylabel('x')
ax.set_xlabel('z')

plt.show()
```

#BIFURCACIÓN HOPF


Para mostrar la ocurrencia de bifurcaciones Hopf en el NCS,
se trazaron los retratos de fase 3D para $a = 0.2, b = 0.1, c = 0$ condiciones iniciales $[0.3, −0.4, −0.1]$ y
diferentes valores de d , Vemos el ciclo límite de bifurcación y se 
convierte a un atractor caótico para $d> d0$ 

```
time = [0,100]
t = np.arange(0,100,0.001)
init_condi = [0.3,-0.4,-0.1]
solucion = solve_ivp(NCS,time,init_condi,args=[0.2,0.1,0,2],t_eval=t)
t = solucion.t
x = solucion.y[0,:]
y = solucion.y[1,:]
z = solucion.y[2,:]
```

```
fig = plt.figure(figsize=(10,8))
fig.suptitle('retratos de fase 3D en el sistema NCS ',fontsize=20)

ax = fig.add_subplot(1, 1, 1, projection='3d')
ax.plot3D(x,y,z,c='k')
ax.set_title('(a)',fontsize=20)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')




plt.show()
```

###DIAGRAMAS DE TRANSICION DE BIFURCACION Y MULTISTABILIDAD


Estudiar el comportamiento dinámico de la NCS con el cambio.
en parámetro, obtenemos e investigamos las gráficas de bifurcación.
Se estudian tres casos diferentes (A, B, C) dependiendo del
elección de parámetros. Las condiciones iniciales se toman como
[0.1, 0.1, 0] y los máximos locales de la variable de estado es
trazado para diferentes valores del parámetro de control para obtener
Los diagramas de bifurcación. 

Caso A: el caso A es el caso 1 en el que elegimos el parámetro
valores $a = 0.2, c = 0$ y $d = 0$, mientras que el parámetro b es
varió de $0$ a $1$. 

El sistema NCS muestra oscilaciones caóticas. El sistema toma un
ruta de duplicación del período al caos en las regiones $0.25 <b <
0.28$ y $0.13 <b <0.15$ la figura  muestra la multiestabilidad.






 
Caso B: los valores de los parámetros para este casos son $b = 0.1, c = 0.1$ y $d = 0$ con parámetro
un ahora que varía de $0.1$ a $0.14$.

Caso C: el caso C es el caso 5 de la tabla 1. Los valores de los parámetros

```
def event_NCS(t,xyz,a,b,c,d):
    x,y,z = xyz
    return x+y+z+1.5
```

```
event_NCS.direction = -1
time = [0,2000]
t = np.linspace(time[0],time[1],10000)
init_condi = [0.1,0.1,0]
```

**Diagrama de Bifurcación de caso A**

```
rs = np.arange(0,1,0.01)
for r in rs:
    solucion = solve_ivp(NCS,time,init_condi,events=[event_NCS],args=[0.2,r,0,0],t_eval=t)
    for i,t_ in enumerate(solucion.t_events[0]):
        if t_ > 20:
            plt.plot(r,solucion.y_events[0][i,1],'.k')
    init_condi = solucion.y[:,-1]
```
