##Modulos de libreria 

```
import matplotlib.style
matplotlib.style.use('classic')
import numpy as np
from matplotlib import pyplot as plt
```

##Funcion metodo de Euler

 es un método iterativo que a partir de un valor inicial, emplea la derivada para conocer la pendiente de la recta tangente a ese punto, para luego trazar una recta hasta el siguiente valor con un incremento h



```
def euler( derivadas, h, tini, tfin, x0, params):
    ti = tini  
    xi = x0
  
    x = list()
    t = list()
  
    while ( ti < tfin ):
    
      x.append (xi)
      
      t.append (ti)
  
      xi = xi + h*derivadas(ti,xi,params)
      ti = ti + h
    return t, x 
```

## Función para método Ruggen-Kutta
es un método iterativo para la aproximación de ecuaciones diferencales ordinarias

```
def rk4( func, h, tini, tfin, x0, params ):
    ti = tini  
    xi = x0
  
    x = list()
    t = list()
  
    while ( ti < tfin ):

      x.append (xi)
      t.append (ti)
      k1 = func( ti,xi,params)
      k2 = func( ti + 0.5*h,xi + 0.5*k1*h,params )
      k3 = func( ti + 0.5*h,xi + 0.5*k2*h,params)
      k4 = func( ti + h,xi + k3*h,params) 

      xi = xi + h*(k1 + 2*k2 + 2*k3 + k4)/6
      ti = ti + h
    return t, x
```

## Ejercicio 2.2.1

SOLUCIÓN DE ECUACIÓN DIFERENCIAL 

\begin{equation}
 \frac{dx}{dt}=(4x^{2}-16)
 \end{equation}

```
def sistema_5(t, x,a):
    x_dot = 4*pow(x,2)-16
    return x_dot
x=np.arange(-10, 10, 0.01)
```

```
x=np.arange(-10, 10, 0.01)

dxdt= 4*pow(x,2)-16
fig = plt.figure()
plt.plot( x, dxdt, linewidth = 2)
plt.title(r'Gráfica $\frac{dx}{dt}$ vs $x$', fontsize=20)
plt.xlabel(r'$x$', fontsize=20)
plt.ylabel(r'$\frac{dx}{dt}$', fontsize=20)
plt.grid()
plt.show()
```

```


a=0
h = 0.015
tini = 0
tfin = 1
x0 = 0

t1,x1 = euler( sistema_5, h, tini, tfin, x0,a)
t2,x2 = rk4( sistema_5, h, tini, tfin, x0,a)


# gráfica de euler:
plt.scatter( t1, x1, c = 'crimson', label = 'Euler', s = 50)
# gráfica de rk4:
plt.scatter( t2, x2, c = 'royalblue', label = 'RK4', s = 50,alpha=0.3)
plt.title('Gráfica de soluciones ecuación diferenial')
plt.xlim([0, tfin])
plt.xlabel(r'$Tiempo$')
plt.grid()
plt.legend(loc='best')
plt.show() 

a=np.arange(-10,3, 0.1)
for i in a:
    x0=i
    t1,x1 = euler( sistema_5, h, tini, tfin, x0,a)
    #plt.plot( t3, x7, c = 'crimson')
    plt.plot( t1, x1, c = 'g')
plt.xlim([0, tfin])
plt.ylim([-6, 7])
plt.xlabel(r'$Tiempo$', fontsize=20)
plt.title('Gráfica de soluciones de la ED con diferentes x0')
plt.grid()
plt.show()
```

 Enla grafica se observa  que las soluciones  tienen dos puntos fijos,Punto Estable en -2 y Punto Inestable en 2
